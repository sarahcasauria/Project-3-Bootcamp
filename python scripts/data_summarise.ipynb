```python
# Import dependencies
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/cleaned_data.csv")

df.head()
```

```python
print(df.Rating.unique())
print(df.Genre.unique())
```

```python
df["E Count"] = np.where(df['Rating']=="E", 1, 0)
df["M Count"] = np.where(df['Rating']=="M", 1, 0)
df["T Count"] = np.where(df['Rating']=="T", 1, 0)
df["E10+ Count"] = np.where(df['Rating']=="E10+", 1, 0)
df["AO Count"] = np.where(df['Rating']=="AO", 1, 0)
df["K-A Count"] = np.where(df['Rating']=="K-A", 1, 0)
df["RP Count"] = np.where(df['Rating']=="RP", 1, 0)
```

```python
df["Sports Count"] = np.where(df['Genre']=="Sports", 1, 0)
df["Racing Count"] = np.where(df['Genre']=="Racing", 1, 0)
df["Platform Count"] = np.where(df['Genre']=="Platform", 1, 0)
df["Misc Count"] = np.where(df['Genre']=="Misc", 1, 0)
df["Action Count"] = np.where(df['Genre']=="Action", 1, 0)
df["Puzzle Count"] = np.where(df['Genre']=="Puzzle", 1, 0)
df["Shooter Count"] = np.where(df['Genre']=="Shooter", 1, 0)
df["Fighting Count"] = np.where(df['Genre']=="Fighting", 1, 0)
df["Simulation Count"] = np.where(df['Genre']=="Simulation", 1, 0)
df["Role-Playing Count"] = np.where(df['Genre']=="Role-Playing", 1, 0)
df["Adventure Count"] = np.where(df['Genre']=="Adventure", 1, 0)
df["Strategy Count"] = np.where(df['Genre']=="Strategy", 1, 0)
```

```python
df.head(100)
```

```python
platform_summary = df.set_index('Platform').groupby(['Platform']).agg({"Year_of_Release": [("earliest_year",'min'), ("latest_year",'max')],
                                             "NA_Sales": [("NA_Sales",'mean')],
                                              "EU_Sales": [("EU_Sales",'mean')],
                                              "JP_Sales": [("JP_Sales",'mean')],
                                              "Other_Sales": [("Other_Sales",'mean')],
                                              "Global_Sales": [("Global_Sales",'mean')],
                                              "Critic_Score": [("mean_Critic_Score",'mean')],
                                              "User_Score" : [("mean_User_Score",'mean')],
                                             "E Count": [("E_Rating_Count",'sum')],
                                              "M Count": [("M_Rating_Count",'sum')],
                                              "T Count": [("T_Rating_Count",'sum')],
                                              "E10+ Count": [("E10_Rating_Count",'sum')],
                                              "AO Count": [("AO_Rating_Count",'sum')],
                                              "K-A Count": [("KA_Rating_Count",'sum')],
                                              "RP Count" : [("RP_Rating_Count",'sum')],
                                             "Sports Count": [("Sports_Genre_Count",'sum')],
                                             "Racing Count": [("Racing_Genre_Count",'sum')],
                                             "Platform Count": [("Platform_Genre_Count",'sum')],
                                             "Misc Count": [("Misc_Genre_Count",'sum')],
                                             "Action Count": [("Action_Genre_Count",'sum')],
                                             "Shooter Count": [("Shooter_Genre_Count",'sum')],
                                             "Puzzle Count": [("Puzzle_Genre_Count",'sum')],
                                             "Fighting Count": [("Fighting_Genre_Count",'sum')],
                                             "Simulation Count": [("Simulation_Genre_Count",'sum')],
                                             "Role-Playing Count": [("RolePlaying_Genre_Count",'sum')],
                                             "Adventure Count": [("Adventure_Genre_Count",'sum')],
                                             "Strategy Count" : [("Strategy_Genre_Count",'sum')]})

platform_summary
```

```python
platform_summary.to_csv("../data/platform_summary.csv")
```
