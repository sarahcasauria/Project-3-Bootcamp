```python
# Import dependencies
import pandas as pd
```

```python
# Read in data from .csv file
data = pd.read_csv("data/Video_Games_Sales_as_at_22_Dec_2016.csv")

# print number of rows
print(len(data))

data.head()
```

```python
# Clean the data by removing all rows that have NaN values
cleaned_data = data.dropna().reset_index()

print(len(cleaned_data))

cleaned_data.head()
```

```python
# Export cleaned data to new csv
cleaned_data.to_csv("data/cleaned_data.csv", index = False)
```
